```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
```

```python
dataset=pd.read_csv("D:\Python Doc\ML Projects\Rock vs Mines  detection for submarine\sonar.csv",header=None)
dataset.head()
```

```python
# number of  rows and columns
dataset.shape
```

```python
dataset.describe()
```

```python
dataset[60].value_counts()   # to check the balance of dataset  M ---->Mines R----->Rock
```

```python
dataset.groupby(60).mean()
```

```python
#splitting the dataset
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,60].values
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Traning the model
log_reg =LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
#evaluate the Model using accuracy score
#acuracy on training data
y_pred=log_reg.predict(x_train)

# score=accuracy_score(y_train,log_reg.predict(x_train))
score=accuracy_score(y_train,y_pred)
print("Accuracy on Training data is:",score)
```

```python
#evaluate the Model using accuracy score
#acuracy on testingdata
y_pred=log_reg.predict(x_test)

# score=accuracy_score(y_train,log_reg.predict(x_train))
score=accuracy_score(y_test,y_pred)
print("Accuracy on Testing data is:",score)
```
